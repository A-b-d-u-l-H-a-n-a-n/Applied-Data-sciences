```python
import pandas as pd
import matplotlib.pyplot as plt

# read data from CSV file
df = pd.read_csv('iris.csv')
```

```python
df.head()
```

```python
df.tail()
```

```python
# select sepal length and species columns
sepal_length = df[['SepalLengthCm', 'Species']]

# group by species and calculate mean sepal length for each year
mean_sepal_length = sepal_length.groupby(['Species']).mean()

# create line graph
mean_sepal_length.plot()
plt.xlabel('Species')
plt.ylabel('Mean Sepal Length')
plt.title('Mean Sepal Length of Iris Species Over Time')
plt.show()
```

```python
sepal = df[['SepalLengthCm', 'SepalWidthCm', 'Species']]

# create scatter plot with different colors for each species
colors = {'Iris-setosa': 'red', 'Iris-versicolor': 'blue', 'Iris-virginica': 'green'}
fig, ax = plt.subplots()
for species, color in colors.items():
    ax.scatter(sepal[sepal['Species'] == species]['SepalLengthCm'], 
               sepal[sepal['Species'] == species]['SepalWidthCm'], 
               c=color, label=species)
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('Sepal Length vs. Sepal Width')
plt.legend()
plt.show()
```

```python
petal_length = df[['PetalLengthCm', 'Species']]

# group by species and calculate mean petal length
mean_petal_length = petal_length.groupby(['Species']).mean()

# create bar plot
mean_petal_length.plot(kind='bar')
plt.xlabel('Species')
plt.ylabel('Mean Petal Length')
plt.title('Mean Petal Length of Iris')
```
